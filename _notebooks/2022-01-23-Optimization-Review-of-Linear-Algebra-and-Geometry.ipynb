# Introduction

The study of optimization can be summed up as the attempt to find the parameters that optimize an objective function. If the objective function describes cost, we may wish to minimize it. Whereas if the objective describes profit, a natural goal would be to maximize it. 

The problems of minimization and maximization, optimization in one word, are the same problem  up to a reflection w.r.t the domain. 

Formally, let the objective function be $f: \mathbb{R^n} \to \mathbb{R}$, and let it have minimizer $x^* \in \mathbb{R^n}$. Then, by definition of minimizer, $f(x^*) \leq f(x) \ \forall x \in \mathbb{R^n}$. It follows that $-f(x^*) \geq -f(x) \forall x \in \mathbb{R^n}$, so $x^*$ is ther maximizer for $-f$.
