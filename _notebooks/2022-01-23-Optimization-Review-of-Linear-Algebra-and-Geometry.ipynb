# Introduction

The study of optimization can be summed up as the attempt to find those parameters that optimize some objective function, if such parameters exist. The objective function can be almost anything â€” cost, profit, nodes in a wireless network, distance to a destination, similarity to a target image, etc. If the objective function describes cost we may wish to minimize it, whereas if the objective describes profit then a natural goal would be to maximize it. 

The problems of minimization and maximization, summed up as *optimization* in one word, are the same problem up to a reflection w.r.t the domain. 

Formally, let the objective function be $f: \mathbb{R^n} \to \mathbb{R}$, and let it have minimizer $x^* \in \mathbb{R^n}$. Then, by definition of minimizer, $f(x^*) \leq f(x) \ \ \forall x \in \mathbb{R^n}$. It follows that $-f(x^*) \geq -f(x) \ \ \forall x \in \mathbb{R^n}$, so $x^*$ is the maximizer for $-f$.
